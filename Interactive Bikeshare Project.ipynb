# US Cities Bikeshare Data


Data from three cities, Chicago, New York City, and Washington D.C have bikesharing programs. All three programs collect data on their bike users to conduct analysis for providing better services to there customers. 
The code cell below provides a method to filter out data by a time frame based on user input and returns various information based on the users choice of time.

*The data was provided by [Motivate](https://www.motivateco.com/)*

```python
import time
import pandas as pd
import numpy as np

CITY_DATA = { 'chicago': 'chicago.csv',
              'new york city': 'new_york_city.csv',
              'washington': 'washington.csv' }



def get_filters():
    
    month_names = ['january', 'febuary', 'march', 'april', 'may', 'june']
    weekday_names = ['monday', 'tuseday','wednesday','thursday','friday','saturday','sunday']
    print('Hello! Let\'s explore some US bikeshare data from 2017!') # get user input for city (chicago, new york city, washington). HINT: Use a while loop to handle invalid inputs
    while True: # lnfinte Loop
        # Takes user's input and saves it as a varaible then breaks from the loop.
        city_input = input('Which city do you want to explore? Chicago, New York City, Washington.\n')
        city_input = city_input.lower()
        if city_input =='chicago':
            city = city_input
            break

        elif city_input == 'new york city':
            city = city_input
            break

        elif city_input == 'washington':
            city = city_input
            break

        else:
            print('Sorry, wrong city. Please try again!')

    while True:
        filter_type = input("Do you want to filter by Month, Day, Both or None?\n")
        filter_type = filter_type.lower()

        if filter_type == 'both':
            while True:
                month_input = input('What months betwen January and June do you want to explore? If you want to see all months, type All.\n')
                month_input = month_input.lower()
                if month_input in month_names:
                    month = month_input
                    break
                elif month_input == 'all':
                    month = month_input
                    break
                else:
                    print("Sorry, month not found")
            while True:
                day_input = input('What day of the week do you want to explore? Type All for all days.\n')
                day_input = day_input.lower()
                if day_input in weekday_names:
                    day = day_input
                    break
                elif day_input == 'all':
                    day == day_input
                    break
                else:
                    print("Sorry, day not found.")
            break

        elif filter_type == 'month':
            
            while True:
                month_input = input('What months betwen January and June do you want to explore? If you want to see all months, type All.\n')
                month_input = month_input.lower()
                if month_input in month_names:
                    month = month_input
                    day = 'all'
                    break
                elif month_input == 'all':
                    month = month_input
                    day = 'all'
                    break
                else:
                    print("Sorry, month not found.")
            break

        elif filter_type == 'day':
            while True:
                day_input = input('What day of the week do you want to explore? Type All for all days.\n')
                day_input = day_input.lower()
                if day_input in weekday_names:
                    day = day_input
                    month = 'all'
                    break
                elif day_input == 'all':
                    day = day_input
                    month = 'all'
                    break
                else:
                    print("Sorry, day not found.")
            break

        elif filter_type == 'none':
            while True:
                if filter_type == 'none':

                    month = month_input = 'all'
                    day = day_input = 'all'
                    break
                else: 
                    print("Sorry, wrong choice")
                    break
            break
    else:
        print("Sorry, wrong choice. Please try again.")
        
    
    print('-'*40)
    return city, month, day


def load_data(city, month, day):
    """
    Loads data for the specified city and filters by month and day if applicable.

    Args:
        (str) city - name of the city to analyze
        (str) month - name of the month to filter by, or "all" to apply no month filter
        (str) day - name of the day of week to filter by, or "all" to apply no day filter
    Returns:
        df - Pandas DataFrame containing city data filtered by month and day
    """
    df = pd.read_csv(CITY_DATA[city])
    df['Start Time'] = pd.to_datetime(df['Start Time']) # loads data and changes 'Start Time' column into datetime.
    df['month'] = df['Start Time'].dt.month # Creates two columns 
    df['day_of_week'] = df['Start Time'].dt.weekday_name
    
    # 
    if month != 'all':
        months = ['january', 'febuary', 'march', 'april', 'may', 'june']
        month =  months.index(month) + 1
        df = df[df['month'] == month]



    if day != 'all':
        df = df[df['day_of_week'] == day.title()]
    
    if day =='all' and month == 'all':
        return df
    
    return df

def time_stats(df):
    

    print('\nCalculating The Most Frequent Times of Travel...\n')
    start_time = time.time()

    # display the most common month
    popular_month = df['month'].mode()[0]

    print("Most popular month:",popular_month   )
    print("\nThis took %s seconds." % (time.time() - start_time))
    print('-'*40)

    


    # display the most common day of week
    popular_week = df['day_of_week'].mode()[0]
    print("Most popular day of the week:", popular_week,)
    print("\nThis took %s seconds." % (time.time() - start_time))
    print('-'*40)


    # display the most common start hour
    df['hour'] = df['Start Time'].dt.hour
    popular_hour = df['hour'].mode()[0]



    print("Most popular hour:", popular_hour )
    print("\nThis took %s seconds." % (time.time() - start_time))
    print('-'*40)
    

def station_stats(df):
    """Displays statistics on the most popular stations and trip."""

    print('\nCalculating The Most Popular Stations and Trip...\n')
    start_time = time.time()

    # display most commonly used start station
    popular_start_station = df['Start Station'].mode()[0]
    start_station_count = df['Start Station'].value_counts().max()

    print("Most commonly used start station:", popular_start_station)
    print("Count:", start_station_count)
    

    # display most commonly used end station
    popular_end_station = df['End Station'].mode()[0]
    end_station_count = df['End Station'].value_counts().max()

    print("Most commonly used end station:", popular_end_station)
    print("Count:", end_station_count)


   


    # display most frequent combination of start station and end station trip

    df['both_station'] = df['Start Station'] + ', ' + df['End Station']
    popular_combo_stations = df['both_station'].mode()[0]
    combo_count = df['both_station'].value_counts().max()



    print("Most commonly used combination of start and end stations:", popular_combo_stations)
    print("Count:", combo_count)
    

    print("\nThis took %s seconds." % (time.time() - start_time))
    print('-'*40)

def trip_duration_stats(df):
    """Displays statistics on the total and average trip duration."""

    print('\nCalculating Trip Duration...\n')
    start_time = time.time()

    # display total travel time
    df['Start Time'] = pd.to_datetime(df['Start Time'])
    df['End Time'] = pd.to_datetime(df['End Time'])

    df['Total Duration'] =  df['End Time'] - df['Start Time'] 
    print('Total time traveled:', df['Total Duration'].sum())


    # display mean travel time
    print("Mean Time Traveled:", df['Total Duration'].mean())
    print("Total Trips:", df['Start Time'].value_counts().sum())


    print("\nThis took %s seconds." % (time.time() - start_time))
    print('-'*40)


def user_stats(df):
    """Displays statistics on bikeshare users."""

    print('\nCalculating User Stats...\n')
    start_time = time.time()


    # Display counts of user types
    user_types = df['User Type'].value_counts()
    print(user_types)

    # Display counts of gender
    if 'Gender' in df:
        gender_types = df['Gender'].value_counts()
        print(gender_types)

    else:
        print('No gender information can be found for this city.')
        
        

    # Display earliest, most recent, and most common year of birth
    if 'Birth Year' in df:

        common_birth = df['Birth Year'].mode()[0]

        earliest_birth = df['Birth Year'].dropna().sort_values().min()

        recent_time = df['Start Time'].sort_values().max()

        most_recent = df['Birth Year'].loc[df['Start Time'] == recent_time]

        print("Most recent birth year", most_recent)
    
        print("Earliest birth year", earliest_birth)

        print("Most common birth year:", common_birth)
    else:
        print('No birth year information can be found for this city.')


    print("\nThis took %s seconds." % (time.time() - start_time))
    print('-'*40)


def raw_data_return(df):
    while True:
        raw_data_input = input('Would you like to see the raw data that you have filtered? yes or no.\n')
        if raw_data_input.lower() == 'yes':
            print(df)
            break
        else:
            break
```

#### By running the cell below you will get a user input question. Please enter in a command.

```python

def main():
    while True:
        city, month, day = get_filters()
        df = load_data(city, month, day)

        time_stats(df)
        station_stats(df)
        trip_duration_stats(df)
        user_stats(df)
        raw_data_return(df)

        restart = input('\nWould you like to restart? Enter yes or no.\n')
        if restart.lower() != 'yes':
            break


if __name__ == "__main__":
    main()
```
